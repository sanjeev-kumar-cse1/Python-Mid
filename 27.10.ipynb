```python
def student_detail(**args):
    return args
student_detail(name='sanjeev',location='Jaipur',batch='A20',email="sa@gmail.com")
```

```python
def student_detail(**args):
    return args
student_detail(name='sanjeev',location='Jaipur',batch='A20',email="sa@gmail.com", adhar=12321)
```

```python
def student_detail(**args):
    for key,value in args.items():
        print(f'{key}:{value}')
student_detail(name='sanjeev',location='Jaipur',batch='A20',email="sa@gmail.com", adhar=112321)
```

```python
x = 20
def change():
    x = 30 # local variable
    return x
print(change())
print(x)
```

```python
x = 20
def change():
    global x # global variable
    x = 30
    return x
change()
print(x)
```

```python
x = 20
def change():
    global x # global variable
    x = 30
    return x
change()
print(x)
print(change())
```
